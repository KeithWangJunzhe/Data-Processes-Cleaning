```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('./megaWordPack/类目热度趋势-面包甜点词包-近365.csv')

df_1.shape
```

```python
print(df_1.columns)
```

```python
#rename
names = ['word','time','c1','c2','c3','exp','click','interact','exp_notes','new_notes',
'ad_exp','ad_click','ad_interact','ad_exp_notes',
'na_exp','na_click','na_interact','na_exp_notes',]
df_1.columns = names
print(df_1.columns)
```

```python
# datetime coversion
df_1['time'] = pd.to_datetime(df_1['time'])
```

```python
# Pivoted table
df_1_p  = df_1.pivot_table(index = 'word', columns = 'time', values = 'click', aggfunc='sum')
df_1_p.head()
```

```python
mega_df= df_1_p
```

```python
#save the file
df_1_p.to_csv('./megaWordPackPivoted/df_1_p.csv', encoding = 'utf_8_sig')
```

```python
Words = list(df_1_p.index) + list(df_2_p.index) + list(df_3_p.index) + list(df_4_p.index) + list(df_5_p.index) +list(df_6_p.index)
```

```python
print(Words)
```

```python
#Insert into the first row
mega_df.insert(0, 'word', Words)
```

```python
#add sum
df_1_p_sum = df_1_p.sum(axis=0, numeric_only= True)
df_2_p_sum = df_2_p.sum(axis=0, numeric_only= True)
df_3_p_sum = df_3_p.sum(axis=0, numeric_only= True)
df_4_p_sum = df_4_p.sum(axis=0, numeric_only= True)
df_5_p_sum = df_5_p.sum(axis=0, numeric_only= True)
df_6_p_sum = df_6_p.sum(axis=0, numeric_only= True)
```

```python
#add to mega_df
mega_df.loc['面包甜品'] = df_1_p_sum
mega_df.loc['速食']= df_2_p_sum
mega_df.loc['零食']= df_3_p_sum
mega_df.loc['非酒精饮品']= df_4_p_sum
mega_df.loc['粮油调味']= df_5_p_sum
mega_df.loc['酒类']= df_6_p_sum
```

```python
mega_df.to_csv('./megaWordPackPivoted/mega_df.csv', encoding = 'utf_8_sig')
```

```python
sum_df.head()
```

```python
time_col = sum_df.columns
time_col.drop('word')
```

```python
sum_df.word = sum_df.word.index[0:6]
```

```python
#melt the dataframe back to long formatt
# drop the word column
sum_long = pd.melt(sum_df, id_vars='word',value_vars = time_col)
sum_long.head()
```

```python
sns.set(font = 'simhei', style = 'whitegrid')
ax = sns.relplot(data = sum_long, x = 'time', y = 'value', hue = 'word', kind = 'line',palette = 'Set2',height = 8, aspect = 1.5,)
ax.set_xticklabels(rotation = 30)
plt.show()
```

```python
#Replot for breakdown
sns.set_theme(style = 'whitegrid',font = 'SimSun')


g = sns.relplot(data = sum_long, x = 'time', y = 'value',col= 'word', hue = 'word', 
                kind = 'line', palette = 'Set2',
                linewidth = 4, zorder = 5, col_wrap = 1, height = 3, aspect = 1.5,
                legend = False)
g.set_xticklabels(rotation = 30)

for time, ax in g.axes_dict.items():
    ax.text(.1, .85, time, transform = ax.transAxes, fontweight = 'bold')
    sns.lineplot(data = sum_long, x = 'time', y = 'value', units = 'word', estimator= None, color = '.7', linewidth = 1, ax = ax)

plt.show()
g.savefig('./TrendBreakdwon.png',dpi = 500)
```
